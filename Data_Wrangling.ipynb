```python
#importing important libraries
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
#to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data.csv")
```

```python
df = df.drop(columns = "Unnamed: 0")
```

```python
df.info()
```

### Data Cleaning

```python
df["milage"] = df["milage"].apply(lambda x: x[0:-6])
```

```python
df["milage"] = df["milage"].str.replace(",", "")
df["milage"] = df["milage"].apply(lambda x: x.strip()).replace("", "0")
df["milage"] = df["milage"].apply(lambda x: int(x))
```

```python
df["engine_size"] = df["engine_size"].str.replace("L", "")
df["engine_size"] = df["engine_size"].apply(lambda x: float(x))
```

```python
df["age"] = df["year"].apply(lambda x: 2025 - x)
df
```

```python
#Removing all values where the milage is 0
df = df[df["milage"] != 0]
```

```python
df["age_scaled"] = MinMaxScaler().fit_transform(df[["age"]])
df["milage_scaled"] = MinMaxScaler().fit_transform(df[["milage"]])
df["engine_size_scaled"] = MinMaxScaler().fit_transform(df[["engine_size"]])
```

```python
df.fuel_type.unique()
```

```python
df["fuel_type_scaled"] = df["fuel_type"].apply(lambda x: 1 if x == "Electric" else 0.8 if x == "Hybrid" else 0.5 if x == "Petrol" else 0.3 if x == "Diesel" else 0)
```

```python
df["age_scaled"] = df["age_scaled"].apply(lambda x: 1 - x)
df["milage_scaled"] = df["milage_scaled"].apply(lambda x: 1 - x)
df["engine_size_scaled"] = df["engine_size_scaled"].apply(lambda x: 1 - x)
df
```

```python
#dropping duplicates
df = df.drop_duplicates()
```

```python
df.to_csv("df.csv")
```
